```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
import pandas as pd

data = {
    'Days': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'],
    'Study_Hours': [8, 6, 7, 5, 7],
    'Marks': [90, 85, 92, 78, 88],
    'Attendance': [95, 90, 92, 88, 95],
    'Sleep_Hours': [7, 6, 8,7,8],
    'Assignments_Completed': [5, 4, 6, 3, 5]
}
df = pd.DataFrame(data)
print(df)
```

```python
#1 Line Chart(Trend)
plt.plot(df['Days'], df['Study_Hours'], marker='o')
plt.title('Study Hours Trend')
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.show()
```

```python
# Advaned line chart
plt.figure(figsize=(10, 6))
plt.plot(df['Days'], df['Study_Hours'], marker='o', color='blue', label='Study Hours')
plt.plot(df['Days'], df['Marks'], marker='o', color='green', label='Marks')
plt.title('Study Hours and Marks Over Time')
plt.xlabel('Days')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
#Bar chart (comparision)
plt.bar(df['Days'], df['Marks'],color= 'skyblue')
plt.title('Marks Comparison')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.show()
```

```python
#Advanced Bar Graph
bars = plt.bar(df['Days'], df['Marks'], color='red')
for bar in bars:
  y = bar.get_height()
  plt.text(bar.get_x() + bar.get_width() / 2, y, str(y),ha='center', va='bottom')
plt.title('Marks Comparison')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.grid(axis='y') # adds horizontal grid for readability
plt.show()
```

```python
#3 Histogram(Distribution)
#Shows how values are spread
data = {
    'Marks': [90, 85, 92, 78, 88,66,34,76,23,45,98,56,89]
}
df= pd.DataFrame(data)
plt.hist(df['Marks'], bins=3, edgecolor='black')
plt.title('Marks Distribution')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()

#
```

```python
#Scatter plot (Relationship)
import matplotlib.pyplot as plt
import pandas as pd

# Re-create the DataFrame with 'Study_Hours' and 'Marks'
data = {
    'Days': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'],
    'Study_Hours': [8, 6, 7, 5, 7],
    'Marks': [90, 85, 92, 78, 88],
    'Attendance': [95, 90, 92, 88, 95],
    'Sleep_Hours': [7, 6, 8,7,8],
    'Assignments_Completed': [5, 4, 6, 3, 5]
}
df = pd.DataFrame(data)

plt.scatter(df['Study_Hours'],df['Marks'])
plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show()
```

```python
#Conditional scatter plot
#Example category
df['Result']= ['Fail','Pass','Fail','Pass','Pass']
colors = ['red' if x=='Fail' else 'green' for x in df['Result']]
plt.scatter(df['Study_Hours'],df['Marks'],c=colors)
plt.title('Study Hours vs Marks(Result-Wise)')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show()
```

```python
#Line plot
sns.lineplot(x='Days',y='Marks', data=df)
plt.title('Seaborn Line Plot')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.show()
```

```python
#Histogram
sns.histplot(df['Marks'],bins=5)
plt.title('Seaborn Histogram')
plt.show()
```
